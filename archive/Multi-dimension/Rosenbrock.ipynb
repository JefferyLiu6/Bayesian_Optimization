```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# NumPy version of the 2D Rosenbrock
def rosenbrock(x, y, a=1.0, b=100.0):
    return (a - x)**2 + b*(y - x**2)**2

# 1. Create a grid on [-5, 10]^2
n_pts = 200
x = np.linspace(-5, 10, n_pts)
y = np.linspace(-5, 10, n_pts)
X, Y = np.meshgrid(x, y)

# 2. Evaluate
Z = rosenbrock(X, Y)

# === Contour plot ===
plt.figure(figsize=(6,5))
cs = plt.contour(X, Y, Z, levels=30)
plt.clabel(cs, inline=True, fontsize=8)
plt.title("Rosenbrock Function Contour")
plt.xlabel("x₁")
plt.ylabel("x₂")

# === 3D surface plot ===
fig = plt.figure(figsize=(7,5))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(
    X, Y, Z,
    rstride=1, cstride=1, alpha=0.9
)
ax.set_title("Rosenbrock Function Surface")
ax.set_xlabel("x₁")
ax.set_ylabel("x₂")
ax.set_zlabel("f(x₁, x₂)")

plt.show()
```
